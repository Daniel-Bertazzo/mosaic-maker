### This notebook aims to show, in details, what the `alter_color` method does to change the tile images' colors to match the canvas image.

```python
import numpy as np
import imageio
from skimage.color import rgb2hsv, hsv2rgb
import sys
import matplotlib.pyplot as plt
from project import downscale, normalize_image
from mpl_toolkits.mplot3d import Axes3D
```

```python
def alter_color(img, predominant_color, min_color, max_color, method=1, normalize=1):
    out_img = np.array(img, copy=True)

    if method == 1:
        predominant_color = predominant_color / np.sum(predominant_color)
        # changes the color channels proportions to the predominant color proportion
        out_img[:, :, 0] = out_img[:, :, 0] * predominant_color[0]  # R
        out_img[:, :, 1] = out_img[:, :, 1] * predominant_color[1]  # G
        out_img[:, :, 2] = out_img[:, :, 2] * predominant_color[2]  # B

    elif method == 2:
        # transforms to the HSV color representation
        out_img = rgb2hsv(out_img)
        predominant_color = rgb2hsv(np.reshape(predominant_color, (1, 1, 3)))

        out_img[:, :, 0] = (predominant_color[:, :, 0])  # alters hue
        out_img[:, :, 1] = (predominant_color[:, :, 1])  # alters saturation

        out_img = hsv2rgb(out_img) * 256

    # returns to original color interval in rgb to accurately represent the intensity (black/white)
    if normalize:
        out_img = normalize_image(out_img, new_min=min_color, new_max=max_color)

    return np.clip(out_img, 0, 255)
```

The cell below does four things:

1) It reads an image and downscales it to a smaller resolution; 

2) Then, It generates a block of solid color to represent a section of the canvas image;

3) Subsequently, It alters its colors using RGB method (with and without normalization);

4) Finally, it alters its colors using HSV method (with and without normalization).

```python
# Reads image and downscales it
img_name = 'grafitti2.jpg'
org_img = imageio.imread('../images/'+img_name)[:, :, 0:3].astype(np.float)

down_img = downscale(org_img, 60, 90).astype(np.uint8)

# Generates a solid block of a given color
# color = np.array([81, 158, 252])
color = np.array([70, 3, 30])

color_block = np.zeros(down_img.shape, dtype=np.uint8)
color_block[:, :, 0] = color[0]
color_block[:, :, 1] = color[1]
color_block[:, :, 2] = color[2]

max_colors = np.zeros((3))
min_colors = np.zeros((3))

max_colors[0] = np.max(down_img[:, :, 0])
max_colors[1] = np.max(down_img[:, :, 1])
max_colors[2] = np.max(down_img[:, :, 2])

min_colors[0] = np.min(down_img[:, :, 0])
min_colors[1] = np.min(down_img[:, :, 1])
min_colors[2] = np.min(down_img[:, :, 2])

# Alters the image's color using RGB method
rgb_norm = alter_color(down_img, color, min_colors, max_colors, method=1).astype(np.uint8)
hsv_norm = alter_color(down_img, color, min_colors, max_colors, method=2).astype(np.uint8)

# Alters the image's color using HSV method
rgb = alter_color(down_img, color, min_colors, max_colors, method=1, normalize=0).astype(np.uint8)
hsv = alter_color(down_img, color, min_colors, max_colors, method=2, normalize=0).astype(np.uint8)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(231); plt.axis('off'); plt.title("Original image"); plt.imshow(down_img)
plt.subplot(232); plt.axis('off'); plt.title("RGB image (not normalized)"); plt.imshow(rgb)
plt.subplot(233); plt.axis('off'); plt.title("HSV image (not normalized)"); plt.imshow(hsv)
plt.subplot(234); plt.axis('off'); plt.title("Solid color used"); plt.imshow(color_block)
plt.subplot(235); plt.axis('off'); plt.title("RGB image (normalized)"); plt.imshow(rgb_norm)
plt.subplot(236); plt.axis('off'); plt.title("HSV image (normalized)"); plt.imshow(hsv_norm)
```

In the cell below, we plot the color distribution of all the images generated above (except the block of solid color). The pixels are represented by the blue dots and the target color is represented by the red triangle. These plots show how the image form (represented by the distribution of the blue dots) behaves in both methods:
* In the RGB method, the image has its colors shifted to be closer to the target color, but still preserving their distribution along the image. This results in an image where we can still see the woman (because the form is preserved) and some shades of the original colors;
* In the HSV method, the image has its colors completely replaced by the solid color's value, but still preserving their distribution along the image. This results in an image where we can also still se the woman, however there is no trace of the original colors, only darker and lighter shades of the target color.

```python
fig = plt.figure(figsize=(16, 20))

# Plots the color distribution of the original downscaled image
ax = fig.add_subplot(321, projection='3d')
plt.title("Original image")
ax.scatter(down_img[:,:,0], down_img[:,:,1], down_img[:,:,2], marker='o', s=5)
ax.scatter(color[0], color[1], color[2], marker='^', s=100, c='r')
ax.set_xlabel('R'); ax.set_ylabel('G'); ax.set_zlabel('B')
ax.set_xlim3d(0, 255); ax.set_ylim3d(0, 255); ax.set_zlim3d(0, 255)

# Plots the color distribution of the RGB method without normalization
bx = fig.add_subplot(323, projection='3d')
plt.title("RGB image (not normalized)")
bx.scatter(rgb[:,:,0], rgb[:,:,1], rgb[:,:,2], marker='o', s=5)
bx.scatter(color[0], color[1], color[2], marker='^', s=100, c='r')
bx.set_xlabel('R'); bx.set_ylabel('G'); bx.set_zlabel('B')
bx.set_xlim3d(0, 175); bx.set_ylim3d(0, 15); bx.set_zlim3d(0, 150)

# Plots the color distribution of the RGB method with normalization
cx = fig.add_subplot(324, projection='3d')
plt.title("RGB image (normalized)")
cx.scatter(rgb_norm[:,:,0], rgb_norm[:,:,1], rgb_norm[:,:,2], marker='o', s=5)
cx.scatter(color[0], color[1], color[2], marker='^', s=100, c='r')
cx.set_xlabel('R'); cx.set_ylabel('G'); cx.set_zlabel('B')
cx.set_xlim3d(0, 255); cx.set_ylim3d(0, 15); cx.set_zlim3d(0, 150)

# Plots the color distribution of the HSV method without normalization
dx = fig.add_subplot(325, projection='3d')
plt.title("HSV image (not normalized)")
dx.scatter(hsv[:,:,0], hsv[:,:,1], hsv[:,:,2], marker='o', s=5)
dx.scatter(color[0], color[1], color[2], marker='^', s=100, c='r')
dx.set_xlabel('R'); dx.set_ylabel('G'); dx.set_zlabel('B')
dx.set_xlim3d(0, 255); dx.set_ylim3d(0, 15); dx.set_zlim3d(0, 150)

# Plots the color distribution of the HSV method with normalization
ex = fig.add_subplot(326, projection='3d')
plt.title("HSV image (normalized)")
ex.scatter(hsv_norm[:,:,0], hsv_norm[:,:,1], hsv_norm[:,:,2], marker='o', s=5)
ex.scatter(color[0], color[1], color[2], marker='^', s=100, c='r')
ex.set_xlabel('R'); ex.set_ylabel('G'); ex.set_zlabel('B')
ex.set_xlim3d(0, 255); ex.set_ylim3d(0, 15); ex.set_zlim3d(0, 150)

plt.show()
```
